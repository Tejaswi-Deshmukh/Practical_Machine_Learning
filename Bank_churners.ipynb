#load the data

```python
df = pd.read_csv("/content/BankChurners.csv")
```

```python
df.isnull().sum()
```

```python
X = df.drop(['Attrition_Flag'], axis=1)
Y = df['Attrition_Flag']
```

```python
#check columns with null values and single unique values
X.shape
```

```python
X.drop_duplicates(inplace=True)
```

```python
X.shape
```

#handle string columns

```python
#apply one hot encoding on all object columns

X_ohe = pd.get_dummies(df)
```

```python
X_ohe.shape
```

```python
X_ohe.columns
```

applying std scaler

```python
X_ohe_std = (X_ohe - X_ohe.mean())/X_ohe.std()
```

```python
X_ohe_std.mean()
```

```python
Y.unique()
```

```python
Y.replace({'Existing Customer':0,
           'Attrited Customer':1},inplace=True)
```

```python
Y.unique()
```

Train Test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train,Y_test = train_test_split(X_ohe_std,Y,
                                    test_size=0.3,
                                    random_state=7)
```

```python
X_train.shape , X_test.shape, Y_train.shape, Y_test.shape
```

#Fine tuning using GridSearch

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

#parameters to be fine tune in SVM
-c : regularization parameter
kernel : 'rbf'
gamma:

```python
param_grid = {
    'C':[10,20,30],
    'kernel':['rbf'],
    'gamma' : [0.001,0.5,1]
}


gs = GridSearchCV(SVC(random_state = 7),param_grid= param_grid,cv=2,verbose=2)
```

```python
gs.fit(X_train,Y_train)
```

```python
gs.best_params_
```

```python
gs.best_score_
```

```python
#Create model with best params

svc = SVC(C=10,gamma=0.001,kernel='rbf',random_state=7)
svc.fit(X_train,Y_train)
Y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_pred))
```

#Fine tuning random forest
-n_estimators[100,500,1000]
-max_features['sqrt','log']
-max_depth[30,50,60]
-oob_score['True']
-max_samples [0.8]

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
par_grid = {
    'n_estimators' : [100,500,1000],
    'max_features' : ['sqrt','log'],
    'max_samples' : [0.8],
    'oob_score' : [True]
}


gs = GridSearchCV(RandomForestClassifier(random_state=7),param_grid= par_grid,cv=2,verbose=2)
```

```python
gs.fit(X_train,Y_train)
```

```python
gs.best_params_
```
