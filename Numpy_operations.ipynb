# Create a sample operation of neuron in numpy

fun neuron has two parameters

x : input array ( size n)

w : weights (size n)

bias : its a single value / constant

x & w are of same size

Step1 : calculate their element wise multiplication

Step2 : calculate sum of above multiplication and add bias to it

Step3 : return the value



```python
def neuron(x,w,bias):
    X = x*w
    X = X.sum() + bias
    return X
```

```python
x= np.array([10,20,30])
```

```python
w=np.random.rand(x.shape[0])
```

```python
w
```

```python
bias = 100
```

```python
neuron(x,w,bias)
```
