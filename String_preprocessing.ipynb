Z-score/Score scaler based Outlier detection

```python
#standardize the data for outlier detection
def outlier_detection_std(df):
  for col in df.columns:
    count_o = df[col].loc[(df[col] < -3) | (df[col] > 3)].shape[0]
    print(col,count_o)
```

```python

outlier_detection_std(X_train_sc[cont_col]) #as outlier detection done only on cont data...hence sending only cont columns of trained data
```

#Outlier handling/outlier imputation using z-score/std scaler
-outlier imputation : replacing the outlier values with the valid values

```python
def outlier_imputation_std(df):
  df1 = df.copy()
  for col in df1.columns:
    df1.loc[(df1[col] < -3) , col] = -3
    df1.loc[(df1[col] > 3) , col] = 3
  return df1
```

```python
X_train_clean = X_train_sc.copy()
X_train_clean[cont_col] = outlier_imputation_std(X_train_sc[cont_col])
```

```python
X_train_clean[cont_col].describe()
```

#Outlier detection using IQR

```python
def outlier_detection_iqr(df):
    for col in df.columns:
        q1,q3 = df[col].quantile([0.25,0.75])
        iqr = q3-q1

        count_o = df[col].loc[(df[col] < (q1-1.5*iqr) )
                            | (df[col] > (q3+1.5*iqr))
                            ].shape[0]
        print(col, count_o)
```

```python
outlier_detection_iqr(X_train_sc[cont_col])
```

```python
def outlier_imputation_iqr(df):
    df1 = df.copy()
    for col in df1.columns:
        q1,q3 = df1[col].quantile([0.25,0.75])
        iqr = q3-q1

        df1.loc[df1[col] < (q1-1.5*iqr)] = (q1-1.5*iqr)
        df1.loc[df1[col] > (q3+1.5*iqr)] = (q3+1.5*iqr)
    return df1
```

```python
X_train_clean1 = X_train_sc.copy()
X_train_clean1[cont_col] = outlier_imputation_iqr(
                            X_train_sc[cont_col])
```
