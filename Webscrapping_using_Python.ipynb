```python
contents
```

parse the content and bring it in a perticular format

```python
soup = BeautifulSoup(contents, "lxml")
```

```python
for tag in soup.findAll(True):
    print(tag)
```

```python
soup.title
```

```python
soup.body
```

```python
for img in soup.findAll("img"):
    print(img['src'])
```
