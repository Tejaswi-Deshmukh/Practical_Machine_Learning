```python
import pandas as pd
```

```python
df = pd.read_csv("/content/sonar_clusters.csv")
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
T = df
```

Model Training and Testing

```python
from sklearn.model_selection import train_test_split
T_train,T_test = train_test_split(T,test_size = 0.4,random_state = 7)
```

```python
T_train.shape,T_test.shape
```

KMeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
mk = KMeans(2,max_iter = 50,random_state = 7)
```

```python
mk.fit(T_train)
```

```python
y_pred = mk.predict(T_test)
```

```python
inertia_values = []
for k in range(2,11):
    km = KMeans(n_clusters = k, random_state=7)
    km.fit(T_train)
    inertia_values.append(km.inertia_)
```

```python
from matplotlib import pyplot as plt
plt.plot(range(2,11), inertia_values)
```

Agglomarative clustering

```python
import matplotlib.pyplot as plt

import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))
plt.title("Dendrograms")
dend = shc.dendrogram(shc.linkage(T.iloc[:100,:], method='ward'))
plt.axhline(y=50, color='r', linestyle='--')
```
