filling the null values with mode values of mode's first value as this can return multimodel values


```python
df.loc[:,cat_col] = df.loc[:,cat_col].fillna(df.loc[:,cat_col].mode().iloc[0])
```

```python
df.loc[:,cont_col] = df.loc[:,cont_col].fillna(df.loc[:,cont_col].median())
```

```python
df.isna().sum()
```

```python
df.loc[:,cont_col] = (df.loc[:,cont_col] - df.loc[:,cont_col].mean()) / df.loc[:,cont_col].std()
```
